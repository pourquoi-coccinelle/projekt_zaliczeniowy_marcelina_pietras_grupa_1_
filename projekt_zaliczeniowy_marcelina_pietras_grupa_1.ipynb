```python
#Ad 6

def remove_stopwords(tekst):
  doc = nlp(tekst)
  result = ""
  for token in doc:
    if token.is_stop == False:
      result += token.text + " "
  return result
      

df_projekt["text_without_stopwords"] = df_projekt["text"].apply(remove_stopwords)
df_projekt
```

CZĘŚĆ 3

```python
#Ad 1

def person(tekst):
  doc = nlp(tekst)
  result = ""
  for ent in doc.ents:
    if ent.label_ == "persName":
      result += ent.text + " "
  return result

df_projekt["persons"] = df_projekt["text"].apply(person)
df_projekt
```

```python
#Ad 2

def place(tekst):
  doc = nlp(tekst)
  result = ""
  for ent in doc.ents:
    if ent.label_ == "placeName":
      result += ent.text + " "
  return result

df_projekt["places"] = df_projekt["text"].apply(place)
df_projekt
```

```python
#Ad 3

def organization(tekst):
  doc = nlp(tekst)
  result = ""
  for ent in doc.ents:
    if ent.label_ == "orgName":
      result += ent.text + " "
  return result

df_projekt["organizations"] = df_projekt["text"].apply(organization)
df_projekt
```

```python
#Ad 4

nlp_blank = spacy.blank("pl")
ruler = nlp_blank.add_pipe("entity_ruler")

pattern = "\S+"
patterns = [{"label": "HASHTAG", "pattern": [{"ORTH": "#"}, {"TEXT": {"REGEX": pattern}}]}]

ruler.add_patterns(patterns)

def hashtag(tekst):
  doc = nlp_blank(tekst)
  result = ""
  for ent in doc.ents:
    if ent.label_ == "HASHTAG":
      result += ent.text + " "
  return result

df_projekt["hashtags"] = df_projekt["text"].apply(hashtag)

df_projekt
```
